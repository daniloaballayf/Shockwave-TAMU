# Convergence Tests

```python
import os
import sys

sys.path.append(os.path.join(os.path.abspath(''), '../'))

import numpy as np
from solver.exact_riem import ext_riem_solver
from solver.euler import laxwendroff, roe, FCT
import matplotlib
import matplotlib.pyplot as plt
```

## Lax-Wendroff

```python
err_rho = []
err_vel = []
err_pre = []
err_ene = []

n_test = np.arange(50, 1001, 50)
C = 0.2
vel_max = 1.5
T = 1
L = 1
t_com = 0.2
x0 = 0.3

for n in n_test:
    # Initial conditions
    dx = L / (n - 1)
    x = np.linspace(0, L, n)
    x0_n = int(x0/dx)
    gamma = 1.4
    u = 0.75*np.ones(n) #np.exp(-(x - 0.5*L)**2 / 0.1)             # Velocity
    u[x0_n:] = 0
    rho = np.ones(n)
    rho[x0_n:] = 0.125
    press_i = np.ones(n)
    press_i[x0_n:] = 0.1
    
    energy = np.zeros(n)
    energy = press_i/((gamma - 1)) + 0.5*rho*u**2
    
    # Parameters 
    dt = C*dx/vel_max
    Nt = int(T // dt)
    idx = int(t_com // dt)
    
    # Analytical Solution
    ext_den, ext_vel, ext_pre, ext_ene = ext_riem_solver(x, x0, rho, u, press_i, energy, dt, Nt)

    # Lax-Wendroff
    sol_rho_lw, sol_mom_lw, sol_pre_lw, sol_ene_lw = laxwendroff(x, rho, u, press_i, energy, dt, Nt)

    # Error 

    err_rho_ = np.linalg.norm(np.abs(sol_rho_lw[:, idx] - ext_den[:, idx]))/np.linalg.norm(ext_den[:, idx])
    err_pre_ = np.linalg.norm(np.abs(sol_pre_lw[:, idx] - ext_pre[:, idx]))/np.linalg.norm(ext_pre[:, idx])
    err_vel_ = np.linalg.norm(np.abs(sol_mom_lw[:,idx]/sol_rho_lw[:,idx] - ext_vel[:, idx]))/np.linalg.norm(ext_vel[:, idx])
    err_ene_ = np.linalg.norm(np.abs((sol_ene_lw[:,idx]/sol_rho_lw[:,idx] - 0.5*(sol_mom_lw[:,idx]/sol_rho_lw[:,idx])**2) - ext_ene[:, idx]))/np.linalg.norm(ext_ene[:, idx])

    err_rho.append(err_rho_)
    err_vel.append(err_vel_)
    err_pre.append(err_pre_)
    err_ene.append(err_ene_)
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

axs[0,0].plot(n_test, err_rho, color='black', linestyle='dotted')
axs[0,0].set_ylabel('Relative Error')
axs[0,0].set_xlabel('Nx')
axs[0,0].set_title('Density')

axs[1,0].plot(n_test, err_pre, color='black', linestyle='dotted')
axs[1,0].set_ylabel('Relative Error')
axs[1,0].set_xlabel('Nx')
axs[1,0].set_title('Pressure')

axs[0,1].plot(n_test, err_vel, color='black', linestyle='dotted')
axs[0,1].set_ylabel('')
axs[0,1].set_xlabel('Nx')
axs[0,1].set_title('Velocity')

axs[1,1].plot(n_test, err_ene, color='black', linestyle='dotted')
axs[1,1].set_ylabel('')
axs[1,1].set_xlabel('Nx')
axs[1,1].set_title('Internal Energy')
plt.show()
```

## Roe
